```python
# q1 : Try to print this by using while loop
n=9
i = 1
while i <= n :
    j = 1
    while j <= i:
        print("*", end = " ")
        j += 1
    print()
    i += 1
```

```python
# q9 : Write a code to send a mail to your friend 
import smtplib, ssl

port = 465  # For SSL
smtp_server = "smtp.gmail.com"
sender_email = "sandipnan206@gmail.com"  # Enter your address
receiver_email = "nandndip220@gmail.com"  # Enter receiver address
password = input("Type your password and Hit enter: ")
message = """\
Subject: Hi there

This message is sent from Ineruron Python Data science Bootcamp."""

context = ssl.create_default_context()
with smtplib.SMTP_SSL(smtp_server, port, context=context) as server:
    server.login(sender_email, password)
    server.sendmail(sender_email, receiver_email, message)
```
